```python
import numpy as np
import pandas as pd
import re
```

```python
with open('exxon_wiki_text.txt','r') as f:
    txt=f.read()
```

```python
txt
```

```python
pattern=re.compile(r'\[\d*\]')
```

```python
txt=pattern.sub(r'',txt)
```

```python
txt=re.sub(r'\n+','\n',txt)
```

```python
import nltk
```

```python
print(txt)
```

```python
from nltk import sent_tokenize
```

```python
import tensorflow as tf
```

```python
from tensorflow.keras.preprocessing.text import Tokenizer
```

```python
tokeniser=Tokenizer()
```

```python
tokeniser.fit_on_texts([txt])
```

```python
tokeniser.word_index
```

```python
sent=sent_tokenize(txt)
```

```python
len(sent)
```

```python
for i in sent:
    print(i,'\n')
```

```python
for sentence in sent:
    print(tokeniser.texts_to_sequences([sentence])[0])
```

```python
[tokeniser.index_word[x] for x in [20, 26, 80, 499, 500, 57, 5, 146, 195, 147, 58, 501, 23, 6] ]
```

```python
input_sequence=[]
for sentence in sent:
    tokenised_sentence=tokeniser.texts_to_sequences([sentence])[0]
    for i in range(1,len(tokenised_sentence)):
        input_sequence.append(tokenised_sentence[:i+1])

input_sequence
```

```python
" ".join([tokeniser.index_word[x] for x in [5,
  30,
  90,
  225,
  343,
  588,
  31,
  166,
  344,
  2,
  10,
  120,
  3,
  94,
  109,
  226,
  128,
  1,
  589,
  590,
  34,
  129]])
```

```python
len(input_sequence)
```

```python
max([len(x) for x in input_sequence])
```

```python
from keras.preprocessing.sequence import pad_sequences
```

```python
padded_input_sequences=pad_sequences(input_sequence)
```

```python
len(padded_input_sequences[1])
```

```python
padded_input_sequences
```

```python
X=padded_input_sequences[:,:-1]
```

```python
X
```

```python
padded_input_sequences[:,-1]
```

```python
y=padded_input_sequences[:,-1]
```

```python
X.shape,y.shape
```

```python
token_count=len(tokeniser.word_index)
token_count
```

```python
from keras.utils import to_categorical
```

```python
y=to_categorical(y,num_classes=token_count+1)
```

```python
y.shape
```

```python
from keras.models import Sequential
```

```python
from keras.layers import Embedding,Dense,LSTM
```

```python
model=Sequential()
model.add(Embedding(y.shape[1],100))
model.add(LSTM(150))
model.add(Dense(y.shape[1],activation='softmax'))
```

```python
model.compile(loss='categorical_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
model.summary()
```

```python
model.fit(X,y,epochs=50)
```

```python
import pickle
```

```python
with open('model_1.pkl','wb')as f:
    pickle.dump(model,f)
```

```python
text='oil'
#tokenize
token_text=tokeniser.texts_to_sequences([text])[0]
#padding
padded_token_text=pad_sequences([token_text],maxlen=X.shape[1],padding='pre')
#prediction
prediction=model.predict(padded_token_text)
```

```python
def predict_text(text):
    for i in range(20):
        text=text
        token_text=tokeniser.texts_to_sequences([text])[0]
        padded_token_text=pad_sequences([token_text],maxlen=X.shape[1],padding='pre')
        prediction=model.predict(padded_token_text,verbose=0)
        text=text+' '+tokeniser.index_word[np.argmax(prediction[0])]
        print(text)
```
